```python
import numpy as np
import matplotlib.pyplot as plt

def generate_wafer_dataset(size, fault_prob):
    """
    Generate a synthetic wafer dataset with potential faults.

    Parameters:
    - size: Size of the square wafer (size x size).
    - fault_prob: Probability of a pixel having a fault.

    Returns:
    - A 2D NumPy array representing the wafer dataset.
    """
    wafer = np.zeros((size, size), dtype=float)

    # Introduce faults randomly based on fault_prob
    faults = np.random.rand(size, size) < fault_prob
    wafer[faults] = 1.0

    return wafer

def plot_wafer(wafer):
    """
    Plot the wafer dataset.

    Parameters:
    - wafer: 2D NumPy array representing the wafer dataset.
    """
    plt.imshow(wafer, cmap='gray', vmin=0, vmax=1)
    plt.title('Wafer Semiconductor Faults')
    plt.colorbar()
    plt.show()

# Example usage
wafer_size = 100
fault_probability = 0.05  # Adjust this value based on desired fault density
wafer_dataset = generate_wafer_dataset(wafer_size, fault_probability)
plot_wafer(wafer_dataset)
```
